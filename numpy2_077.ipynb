```python
import numpy as np
arr = np.array([[1,2,0,4],[4,-0.5,6,0],[2,6,7,8],[3,-7,4,2]])
print("Original array:\n", arr)
```

```python
print("Every other rows:\n", arr[::2])
```

```python
arr = np.array([1,2,3,4,5,6,7])
print("Original array:", arr)
print("Return every other element in the array :", arr[::2])
```

```python
arr = np.array([[1,2,0,4],[4,-0.5,6,0],[2,6,7,8],[3,-7,4,2]])
temp = arr[:2, :3]
print("\nArray with first 2 rows and 3 columns:\n", temp)
```

```python
newarr = np.array([[0,1,2,3],[3,2,1,0]])
print("\nElements at indices(0,3),(1,2),(2,1),(3,0):\n", newarr)
```

```python
cond = arr > 2
temp = arr[cond]
print("\nElements greater than 2:\n", temp)
```

```python
import numpy as np
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr = np.concatenate((arr1, arr2))
print(arr)
```

```python
arr3 = np.hstack((arr1, arr2))
print(arr3)
```

```python
arr4 = np.vstack((arr1, arr2))
print(arr4)
```

```python
arr5 = np.dstack((arr1, arr2))
print(arr5)
```

```python
arr = np.array([1,2,3,4,5,6])
arr6 = np.array_split(arr, 3)
print(arr)
print(arr6)
print(arr6[0])
print(arr6[1])
print(arr6[2])
```
